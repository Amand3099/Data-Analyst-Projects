## Описание проекта

**Заказчик исследования** — отдел маркетинга банка «Метанпром». Банк располагается в Ярославле и областных городах: Ростов Великий и Рыбинск. В банке за последнее время существенно повысился отток клиентов. Это начинает становиться проблемой, которая угрожает существованию банка, поскольку привлекать новых клиентов дороже, чем удерживать уже имеющихся. 

**Цель исследования** — выделить сегменты клиентов, склонных к уходу из банка, и подготовить план действий по снижению оттока и повышению удержания клиентов.

- Осуществить предобработку данных;
- Провести исследовательский анализ данных (EDA);
- Проверить гипотезы;
- Определить стратегическое влияние показателей на отток;
- Произвести сегментацию на основе стратегических показателей;
- Сформулировать итоговый вывод и разработать рекомендации по уменьшению оттока клиентов.

#### Поиск и устранение пропусков

```python
df.loc[df['age'].isna()]
```

Ранее, были обнаружены пропуски в столбцах 'age' и 'balance'. В первом случае восстановить данные не получится, поскольку невозможно верно "угадать" возраст. Количество пропусков в столбце 'age' весьма несущественно (26 или 0.26%). Возможная причина - ошибки при записи данных. Учитывая крайне малое количество, пропуски можно удалить без особых потерь.

```python
df.dropna(subset=['age'], inplace=True) 
df.info()
```

Пропусков в столбце 'balance' существенно больше (2279 или 22.8%). Перед тем, как предпринять какие-либо меры по их устранению, ознакомимся с ними более подробно.

```python
df.loc[df['balance'].isna()]
```

Примечательно, что у всех наблюдаемых строк с пропусками в 'balance', значение 'churn' 0. Удостоверимся, нет ли среди них уходящих.

```python
df.loc[(df['balance'].isna()) & (df['churn'] == 1)]
```

Действительно, среди этих клиентов есть отточные. Тогда, проверим, есть ли пользователи с балансом равным нулю, поскольку в данном случае NaN может и быть нулем.

```python
df.loc[df['balance'] == 0]
```

Клиенты с балансом, равным нулю - есть, хотя их и всего двое. Возможно, пропуски в 'balance' означают, что клиенты пользуются услугами банка, не открывая лицевой счет. Исследуем пропуски на предмет зависимостей и закономерностей и сравним их с остальными данными. 

```python
print('Все данные')
print()
for column in ['city','products','credit_card','last_activity', 'churn']:
    print(column)
    print(round(df[column].value_counts(normalize=True),2))
    print()
   
print('Данные без пропусков')
print()
no_omissions = df.loc[df['balance'].isna() == False]
for column in ['city','products','credit_card','last_activity', 'churn']:
    print(column)
    print(round(no_omissions[column].value_counts(normalize=True),2))
    print()
    
print('Данные с пропусками')
print()
omissions = df.query('balance.isna()')
for column in ['city','products','credit_card','last_activity', 'churn']:
    print(column)
    print(round(omissions[column].value_counts(normalize=True),2))
    print()  
```

Больше всего клиентов с пропусками в балансе - из Ярославля. Однако, это совпадает с распределением клиентов по городам в целом (в т.ч. для данных без пропусков), поэтому нельзя утверждать, что здесь есть зависимость. Среди таких клиентов практически все пользуются 1 продуктом, нет тех кто пользовался бы 4 и более. Это отличает строки с пропусками от строк без пропусков: на первом месте с большим отрывом клиенты, пользующиеся 2 продуктами; есть клиенты с 4 продуктами.  В основном пропуски в балансе у клиентов с кредитной картой, однако это также не свидетельствует о наличии какой-либо связи, поскольку клиентов с кредитными картами больше впринципе. Признак недавней активности также несильно отличается от клиентов без пропусков в балансе. Примечательно, что отточных среди клиентов с пропусками практически нет (1%), в отличие от строк без пропусков: отточных клиентов около 23%. Тем не менее, и в этом случае пока что рано однозначно утверждать о наличии взаимосвязи.

Изучим распределение по рейтингу, возрасту, заработной плате и количеству объектов в собственности.

```python
print('Строки без пропусков')
for column in ['age','equity', 'est_salary', 'score']:
    no_omissions.plot(y = column, kind = 'hist', bins = 30, grid = True)
    plt.ylabel('Количество клиентов')
```

```python
print('Строки с пропусками')
for column in ['age','equity', 'est_salary', 'score']:
    omissions.plot(y = column, kind = 'hist', bins = 30, grid = True)
    plt.ylabel('Количество клиентов')
```

В распределении по возрасту серьезных различий не выявлено. Есть различие в распределении, по количеству объектов собственности: в строках без пропусков клиентов с 0 объектов достаточно мало, в то время как в строках с пропусками - наоборот, у большинства 0, практически нет никого даже с 1 объектом. По распределению з/п строки с пропусками выходят вперед: "пик" смещен ближе к 200 тыс., больше клиентов с з/п >400 тыс. Очень различаются графики распределений по параметру 'score': у строк без пропусков основная масса сосредоточена в районе ~800-900 баллов, в то время, как у строк с пропусками на этом месте "яма", разделяющаяя относительно небольшое число клиентов со скорингом 700-750 и большинство клиентов со скорингом 900-925 баллов.

Дополнительно, посмотрим на средние значения признаков в разбивке по наличию пропусков:

```python
omissions.mean() # средние по строкам с пропусками
```

```python
no_omissions.mean() # средние по строкам без пропусков
```

```python
omissions.mean() - no_omissions.mean()  # разница средних
```

В целом, можно сказать, что наблюдается ряд различий по средним. У клиентов с пропусками средний балл скоринга выше на 21, по возрасту различий почти нет. В среднем, у клиентов с пропусками
нет объектов собственности, в то время как у клиентов без пропусков среднее значение около 3. Клиенты без пропуска в балансе в среднем используют на 1 продукт больше (2). Среди клиентов с пропусками больше тех, у кого есть кредитная карта (81%), чем среди клиентов без пропусков (64%). По недавней активности обе группы примерно равны. Средняя оценочная зарплата у клиентов с пропусками выше на 101 тыс. и в этой группе практически нулевой отток.


Судя по всему, пропуски в балансе имеют закономерность: у таких клиентов в большинстве случаев высокий скоринг и з/п, 0 объектов собственности, есть кредитная карта, пользуются 1 продуктом и являются максимально неотточными (отток менее 1%). Категория - MAR. Вероятно, это действительно клиенты, которые попросту не имеют лицевого счета, поэтому и баланс отсутствует. Их можно выделить в отдельную категорию - "Клиенты без лицевого счета". Учитывая практически нулевой отток и ярко выраженные отличия от большинства клиентов по ряду признаков, их следует рассматривать отдельно. Поскольку цель всего исследования - исследование оттока и выявление самых проблемных сегментов, указанный пласт клиентов может сильно искажать общую картину. Таким образом, данные об этих клиентах следует перенести в отдельный датафрейм. Хотя "Клиенты без лицевого счета" и не будут выделены в отдельный сегмент ввиду нулевой отточности, в конце исследования будет сформулирована дополнительная рекомендация с учетом исключительности ряда показателей данной категории.

```python
no_balance = df.loc[df['balance'].isna() == True]
df.dropna(subset=['balance'], inplace=True) 
```

- На графике баланса наглядно видно, что присутствует множество аномально высоких значений. Визуально, дать оценку очень трудно. Требуется более тщательное исследование.

- Среднее значение по оценочной заработной плате - в районе ~140-150 тыс. Присутствует большое количество выбросов в большую сторону. На выбросы похожи все значения больше ~300 тыс. Их также потребуется изучить.

## Проверка гипотез

### Гипотеза №1

Проверим, есть ли статистически значимое различие в уровне доходов между клиентами, ушедшими в отток, и клиентами, которые продолжают пользоваться услугами банка.

Н0 - статистически значимых различий в средних выборок доходов между клиентами, ушедшими в отток, и клиентами, которые продолжают пользоваться услугами банка нет.

Н1 - статистически значимые различия в средних выборок доходов между клиентами, ушедшими в отток, и клиентами, которые продолжают пользоваться услугами банка есть.

```python
plt.figure(figsize=(12, 7))
sns.histplot(data=df, x='est_salary', hue='churn')
plt.title('Распределение клиентов по уровню оценочнной зарплаты')
plt.xlabel('Оценочная зарплата, ед.')
plt.ylabel('Количество клиентов, чел.')
plt.show()
```

Создадим две выборки - те, кто является клиентом банка, и те, кто покинул.

```python
salary_churn_0 = df.query('churn == 0')['est_salary']
salary_churn_1= df.query('churn == 1')['est_salary']
```

Посмоторим на показатели по выборкам.

```python
salary_churn_0.describe()
```

```python
salary_churn_1.describe()
```

Мы видим, что показатели среднего и стандартного отклонения по выборкам приблизительно равны. 

Также проверим показатели дисперсии.

```python
print('Дисперсия клиентов не ушедших в отток:', np.var(salary_churn_0, ddof=1))
print('Стандартное отклонение:', np.sqrt(np.var(salary_churn_0, ddof=1)))
print('')
print('Дисперсия клиентов ушедших в отток:', np.var(salary_churn_1, ddof=1))
print('Стандартное отклонение:', np.sqrt(np.var(salary_churn_1, ddof=1)))
```

Показатели дисперсии для выборок относительно близки.

Сначала проверим гипотезу c помощью t-критерия Стьюдента. Учтем, что выборки обладают достаточной величиной данных, в разы различаются по объемам и независимы друг от друга. Применять будем метод ttest_ind библиотеки scipy для сравнения средних двух независимых выборок с различной дисперсией.

```python
#За критический уровень статистической значимости alpha примем значение 0,05
alpha = 0.05
#Выполним t-тест
results_salary = st.ttest_ind(salary_churn_0, salary_churn_1, equal_var=False)
print('p-значение:', results_salary.pvalue)

if results_salary.pvalue<alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

Нулевая гипотеза была отвергнута, а значит, по результатам проверки с помощью t-критерия Стьюдента статистически значимые различия в уровне доходов между клиентами, ушедшими в отток, и клиентами, которые продолжают пользоваться услугами банка - есть.

Далее проверим гипотезу непарметрическим тестом с помощью критерия Манна-Уитни. U-критерий Манна-Уитни является непараметрическим критерием, поэтому, в отличие от t-критерия Стьюдента, не требует наличия нормального распределения сравниваемых совокупностей. 

Н0 - статистически значимых различий в распределении уровня доходов между клиентами, ушедшими в отток, и клиентами, которые продолжают пользоваться услугами банка - нет.

Н1 - статистически значимые различия в распределении уровня доходов между клиентами, ушедшими в отток, и клиентами, которые продолжают пользоваться услугами банка - есть.

```python
#За критический уровень статистической значимости alpha примем значение 0,05
alpha = 0.05
#Выполним тест Манна-Уитни
results_salary = st.mannwhitneyu(salary_churn_0, salary_churn_1, alternative='two-sided')
print('p-значение:', results_salary.pvalue)

if results_salary.pvalue<alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

Нулевая гипотеза оказалась отвергнута. Следовательно, по результатам проверки с помощью u-критерия Манна-Уитни статистически значимые различия в уровне доходов между клиентами, ушедшими в отток, и клиентами, которые продолжают пользоваться услугами банка - есть.

Таким образом, результаты обоих тестов совпали. ### Гипотеза №2

Проверим, есть ли статистически значимое различие в количестве используемых продуктов банка между клиентами, ушедшими в отток, и клиентами, которые продолжают пользоваться услугами банка.

Н0 - статистически значимых различий в средних выборок количества используемых продуктов банка между клиентами, ушедшими в отток, и клиентами, которые продолжают пользоваться услугами банка нет.

Н1 - статистически значимые различия в средних выборок количества используемых продуктов банка между клиентами, ушедшими в отток, и клиентами, которые продолжают пользоваться услугами банка есть.

```python
plt.figure(figsize=(10, 7))
sns.countplot(x='products', hue='churn', data=df)
plt.title('Распределение клиентов по количеству используемых продуктов')
plt.xlabel('Количество используемых продуктов, шт.')
plt.ylabel('Количество клиентов, чел.')
plt.show()
```

Создадим две выборки - те, кто является клиентом банка, и те, кто покинул.

```python
prod_churn_0 = df.query('churn == 0')['products']
prod_churn_1= df.query('churn == 1')['products']
```

Посмоторим на показатели по выборкам.

```python
prod_churn_0.describe()
```

```python
prod_churn_1.describe()
```

Мы видим, что показатели среднего и стандартные отклонения по выборкам близки

Также проверим показатели дисперсии.

```python
print('Дисперсия клиентов не ушедших в отток:', np.var(prod_churn_0, ddof=1))
print('Стандартное отклонение:', np.sqrt(np.var(prod_churn_0, ddof=1)))
print('')
print('Дисперсия клиентов ушедших в отток:', np.var(prod_churn_1, ddof=1))
print('Стандартное отклонение:', np.sqrt(np.var(prod_churn_1, ddof=1)))
```

Показатели дисперсии выборок имеют значительные различия.

Проверять гипотезу будем c помощью u-критерия Манна-Уитни, поскольку одним из ключевых условий для проведения t-теста является равенство дисперсий. 

```python
#За критический уровень статистической значимости alpha примем значение 0,05
alpha = 0.05
#Выполним тест Манна-Уитни
results_prod = st.mannwhitneyu(prod_churn_0, prod_churn_1, alternative='two-sided')
print('p-значение:', results_prod.pvalue)

if results_prod.pvalue<alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

Нулевая гипотеза оказалась отвергнута. Значит, статистически значимые различия в количестве используемых продуктов банка между клиентами, ушедшими в отток, и клиентами, которые продолжают пользоваться услугами банка - есть.

Проведем дополнительную проверку с помощью t-критерия Стьюдента.

```python
alpha = 0.05
results_prod = st.ttest_ind(prod_churn_0, prod_churn_1, equal_var=False)
print('p-значение:', results_prod.pvalue)

if results_prod.pvalue<alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

В данном случае, результаты обоих тестов совпадают. Если бы возникло расхождение результатов, более надежным следовало бы считать u-критерий Манна-Уитни.

В соответствии с поставленной целью и задачами, были сформулированы следующие **рекомендации**:

**Сегмент 1**: является вторым по численности, имеет самый низкий процент оттока. Однако, это сегмент vip-клиентов. Стоит рассмотреть возможность реализации программы лояльности премиум класса. Существующие в настоящее время программы лояльности сводятся к одному из следующих видов:

- изменение процентной ставки или комиссии (увеличение ставки по депозиту, снижение тарифа по РКО или ставки за пользование ссудой);

- предоставление льготного периода пользования продуктом;

- предоставление бесплатных бонусных продуктов (например, БМБ-банкинга по корпоративной карте) и начисление бонусных баллов;

- скидки на услуги и товары в сети партнеров банка - членов корпоративного клуба и др.

Учитывая состоятельность клиентов данного сегмента, наиболее перспективным представляется вариант изменения процентной ставки или комиссии. Тем не менее, стоит рассмотреть возможность внедрения и других вариантов из списка с поправкой на финансовое состояние этих пользователей.


**Сегмент 2**: первый по численности, поэтому, несмотря на относительно небольшой процент оттока, ему следует уделить внимание. Поскольку, в данном сегменте кредитной карты нет у большой части клиентов, рекомендуется повысить количество клиентов с кредитной картой. Сделать это можно, например, с помощью контекстной или таргетированной рекламы. Продвижение кредитной карты, в целом, актуально для всех сегментов. Однако, именно для сегмента 2 эта мера может быть наиболее действенна ввиду количества пользователей в данном сегменте. Также, стоит рассмотреть возможность проведения разного рода акций для повышения лояльности среди мужчин (например, скидки на услуги и товары в сети партнеров банка - членов корпоративного клуба и др).
 

**Сегмент 3**: имеет самый высокий процент оттока. Рекомендуются следующие меры: 


1) Ввиду того, что среди отточных клиентов много тех, кто был недавно активен, требуется пересмотреть внедренные за последнее время продукты. Вероятно, отказ или переработка части из них сможет снизить процент уходящих клиентов.


2) Для снижения оттока в 3 сегменте рекомендуется оптимизировать предлагаемые продукты: определить популярные продукты, объединить некоторые продукты в один, отказаться от непопулярных.

Среди рекомендуемых вариантов продуктов:

- страховые, брокерские, инвестиционные продукты с использованием онлайн приложения или мобильного приложения с возможностью не открывать лицевой счет;

- услуги по переводам, оплате счетов для физических лиц и малого бизнеса.

- услуги для самозанятых.
